```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/pokemon_data.csv')
```

```python
df = df.drop(columns='Generation')
df = df.drop(columns='#')
```

```python
df['Type 1'].unique()
```

```python
df[df.Name == 'Pikachu']
```

```python
dfRock = df.loc[(df['Type 1'] == 'Rock') | (df['Type 2'] == 'Rock')]
```

```python
dfRock = dfRock[(df['Type 1'] != 'Flying') & (df['Type 2'] != 'Flying')]
dfRock = dfRock[(df['Type 1'] != 'Water') & (df['Type 2'] != 'Water')]
```

```python
dfRock = dfRock.loc[(df['Attack'] > 40) & (df['Defense'] > 55)]
dfRock = dfRock.loc[(df['Sp. Atk'] > 50) & (df['Sp. Def'] > 50)]
```
